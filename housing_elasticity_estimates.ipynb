```python
from pathlib import Path
import pandas as pd

housing_supply_path = Path("data") / "HOUSING_SUPPLY (2).dta"

housing_supply = pd.read_stata(housing_supply_path)

print(housing_supply.shape)
housing_supply.head()
```

```python
state_patterns = {"Texas": ", TX", "Arizona": ", AZ"}

elasticity_summary = []
for state, pattern in state_patterns.items():
    state_df = housing_supply[housing_supply["msaname"].str.contains(pattern, regex=False, na=False)]
    elasticity_summary.append({
        "state": state,
        "metro_count": len(state_df),
        "average_elasticity": state_df["elasticity"].mean(),
        "population_weighted_average_elasticity": (
            state_df["elasticity"] * state_df["population"]
        ).sum() / state_df["population"].sum(),
    })

elasticity_summary = pd.DataFrame(elasticity_summary)
print(elasticity_summary.to_string(index=False))
```
