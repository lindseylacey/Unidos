```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\linds\OneDrive\Documents\CII2 Multi-site\City_of_Seattle_Wage_Data_20260226.csv")

#13,981 observations
```

```python
df['JOB_TITLE_STD'] = (
    df['Job Title']
    .fillna('')
    .astype(str)
    .str.lower()
    .str.replace(r'[^a-z0-9\s]', ' ', regex=True)
    .str.replace(r'\b(sr|snr)\b', 'senior', regex=True)
    .str.replace(r'\bjr\b', 'junior', regex=True)
    .str.replace(r'\s+', ' ', regex=True)
    .str.strip()
)

top_titles = df.loc[df['JOB_TITLE_STD'] != '', 'JOB_TITLE_STD'].value_counts().head(15)

plt.figure(figsize=(10, 6))
plt.bar(top_titles.index.str.title(), top_titles.values)
plt.xlabel('Standardized Job Title')
plt.ylabel('Count')
plt.title('Top 15 Standardized Job Titles')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
hourly_rate = pd.to_numeric(
    df['Hourly Rate '].astype(str).str.replace(r'[$,]', '', regex=True),
    errors='coerce'
).dropna()

plt.figure(figsize=(10, 6))
counts, bin_edges, _ = plt.hist(hourly_rate, bins=20, edgecolor='black')
plt.xticks(bin_edges, [f'{edge:,.2f}' for edge in bin_edges], rotation=45, ha='right')
plt.xlabel('Hourly Rate (USD)')
plt.ylabel('Frequency')
plt.title('Distribution of Hourly Rate')
plt.tight_layout()
```
